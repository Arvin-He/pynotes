#### 去除相邻相同元素,只保留一个

给出一个列表,输出去除相邻相同元素,且只保留一个.

输入: aa = [1, 3, 5, 5, 5, 3, 2, 1, 1, 2, 3, 3]

输出: bb = [1, 3, 5, 3, 2, 1, 2, 3]

```python
aa = [1, 3, 5, 5, 5, 3, 2, 1, 1, 2, 3, 3]
print(len(aa))

bb = [item for i, item in enumerate(aa[1:]) if aa[i]!=aa[i+1] and len(aa)>1]
bb.insert(0, aa[0])
print(bb)
```

### 展开嵌套列表

```python
a = [[1,2,3], [4, 5, 6], [7, 8]]

def flatten(items):
    res = []
    for item in items:
        res.extend(item)
    return res

print(flatten(a))

# 生成器
from collections import Iterable

def flatten(items, ignore_types=(str, bytes)):
    for x in items:
        if isinstance(x, Iterable) and not isinstance(x, ignore_types):
            yield from flatten(x)
        else:
            yield x
```

### 判断是否是闰年

```python
def is_leap_year(year):
    if (year % 4) == 0 and (year % 100) != 0 or (year % 400) == 0:
        return True
    else:
        return False
```

### 深拷贝与浅拷贝

```python
import copy

a = [1, {'aa': 123}, 5]

print('浅拷贝')
b = copy.copy(a)
print(b)

print('深拷贝')
c = copy.deepcopy(a)
print(c)

print('改变原始列表中对象的值')
a[1]['aa'] = 22
print(a)

print('再次查看浅拷贝的值')
print(b)

print('再次查看深拷贝的值')
print(c)
```

```python
import copy
from pprint import pprint


curr_list = [{'port': 22, 'product': 'ssh'},
    {'port': 24, 'product': 'http'},
    {'port': 26, 'product': 'tcp'},
    {'port': 28, 'product': 'mail'},
    {'port': 30, 'product': 'rpc'}]

old_list = [{'port': 22, 'product': 'ssh'},
    {'port': 23, 'product': 'http'},
    {'port': 24, 'product': 'http'},
    {'port': 26, 'product': 'tcp'},
    {'port': 28, 'product': 'udp'},
    {'port': 30, 'product': 'rpc'},
    {'port': 32, 'product': 'mail'}]

copy_data_list = copy.deepcopy(curr_list)

for data in old_list:
    port = data.get('port')
    product = data.get('product')
    for item in copy_data_list:
        if port == item.get('port') and product == item.get('product'):
            curr_list.remove(item)
        
pprint(curr_list)        
```
