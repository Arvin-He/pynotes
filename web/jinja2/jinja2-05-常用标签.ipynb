{{  }} 变量替换

循环
```
{% for user in users %}
    {%- if loop.index is even %}{% continue %}{% endif %}
    ...
{% endfor %}
```

{% if expresssion %} {% elif expression %} { % else % } {% endif %}

###  全局函数清单

range([start, ]stop[, step])

lipsum(n=5, html=True, min=20, max=100) 

在模板中生成 lorem ipsum 乱数假文。默认会生成 5 段 HTML ，每段在 20 到 100 词之间。如果 HTML 被禁用，会返回常规文本。测试布局时生成简单内容时很有 用。

dict(**items)

class cycler(*items) 周期计允许你在若干个值中循环
类似 loop.cycle 的工作方式。不同于 loop.cycle 的是，无论如何你都可以在循环外或在多重循环中使用它。

周期计有下面的属性和方法:
```
reset() 重置周期计到第一个项。
next() 返回当前项并跳转到下一个。
current 返回当前项。
```
